```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mse(y, pred):
  return ((y - pred)**2).mean()
```

```python
def get_dataset(N):
    x = np.random.uniform(low=0.0, high=2 * math.pi, size=N)
    return x, np.sin(x)
```

```python
x_train, y_train = get_dataset(100)
```

```python
plt.plot(x_train, y_train, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['targets (labels)'])
plt.show()
```

```python
predictions = 0.8 * np.sin(x_train)
mse = mse(y_train, predictions)
```
